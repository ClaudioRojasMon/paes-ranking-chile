```python
df = pd.read_csv(PAES_2023, sep=';', low_memory=False)
```

```python
# b) Calcular promedios por RBD para estudiantes con SITUACION_EGRESO = 1
filtered_data = df[df['SITUACION_EGRESO'] == 1]
columns_to_average = ['CLEC_REG_ACTUAL', 'MATE1_REG_ACTUAL', 'MATE2_REG_ACTUAL', 'HCSOC_REG_ACTUAL', 'CIEN_REG_ACTUAL']
rbd_averages = filtered_data.groupby('RBD')[columns_to_average].mean().reset_index()
```

```python
# c) Calcular el promedio de CLEC_REG_ACTUAL y MATE1_REG_ACTUAL por RBD
columns_paes = ['CLEC_REG_ACTUAL', 'MATE1_REG_ACTUAL']
rbd_paes_average = filtered_data.groupby('RBD')[columns_paes].mean()
rbd_paes_average['PAES_PROMEDIO'] = rbd_paes_average.mean(axis=1)
rbd_paes_average = rbd_paes_average.reset_index()
```

```python
# d) Crear un ranking de colegios basado en PAES_PROMEDIO
rbd_paes_average['RANK'] = rbd_paes_average['PAES_PROMEDIO'].rank(ascending=False, method='min')
rbd_paes_average = rbd_paes_average.sort_values('PAES_PROMEDIO', ascending=False).reset_index(drop=True)
```

```python
# Función para consultar el lugar de un colegio dado su RBD
def obtener_posicion(rbd):
    resultado = rbd_paes_average[rbd_paes_average['RBD'] == rbd]
    if resultado.empty:
        return f"El RBD {rbd} no se encuentra en el ranking."
    else:
        return resultado[['RBD', 'PAES_PROMEDIO', 'RANK']]
```

```python
# Mostrar resultados
print("Promedios obtenidos por estudiantes para cada RBD:\n", rbd_averages)
print("\nPromedio PAES por RBD:\n", rbd_paes_average)
```

```python
# Ejemplo de consulta de ranking
rbd = int(input("\nIngrese el RBD para consultar el ranking: "))
print(obtener_posicion(rbd))
```
